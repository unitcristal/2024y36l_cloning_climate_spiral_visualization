```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data setup (a small part of your dataset)
data = {
    'Year': [1880, 1880, 1880, 1880, 1880, 1881, 1881, 1881],
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jan', 'Feb', 'Mar'],
    'Anomaly': [-0.2, -0.25, -0.09, -0.16, -0.09, -0.19, -0.15, 0.02]
}

# Convert the data into a DataFrame
df = pd.DataFrame(data)

# Combining Year and Month to create a time axis
df['Date'] = pd.to_datetime(df['Year'].astype(str) + df['Month'], format='%Y%b')

# Plotting
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Anomaly'], marker='o', linestyle='-', color='b')
plt.title('Temperature Anomalies Over Time')
plt.xlabel('Date')
plt.ylabel('Anomaly')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
